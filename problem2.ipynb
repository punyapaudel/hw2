# a
Code to find solution of $ùë•^{3} ‚àí 7ùë•^{2} + 14ùë• ‚àí 5 = 0$ to a precision of $\epsilon < 10^{-8}$.

```python
#BISECTION METHOD
#equation
def f(x):
    return x**3 - 7*x**2 + 14*x - 5

#initial guess
a = 0.0
b = 1.0
c_old = 1.0  # Placeholder to start
step = 0
rel_error = 1.0

while rel_error >= 1e-8:
    step += 1
    c = (a + b) / 2  # Find the midpoint
    
    # Calculate error (skipping the very first step)
    if step > 1:
        rel_error = abs((c - c_old) / c)
        
    print(f"Step {step}: x = {c:.8f}, Error = {rel_error:.8e}")
    
    # Update boundaries for the next loop
    if f(a) * f(c) < 0:
        b = c  # Root is in the left half
    else:
        a = c  # Root is in the right half
        
    c_old = c  # Save current midpoint for the next error calculation

print(f"\nFinal Answer: x = {c:.8f} found in {step} steps.")
```

```python
#NEWTON-RAPHSON METHOD
#equation
def f(x):
    return x**3 - 7*x**2 + 14*x - 5

def f_prime(x):
    # The derivative: 3x^2 - 14x + 14
    return 3*x**2 - 14*x + 14

x = 0.0  # initial guess
step = 0
rel_error = 1.0

while rel_error >= 1e-8:
    step += 1
    
    # Calculate the next guess using the Newton-Raphson formula
    x_new = x - (f(x) / f_prime(x))
    
    # Calculate the relative error
    rel_error = abs((x_new - x) / x_new)
        
    print(f"Step {step}: x = {x_new:.8f}, Error = {rel_error:.8e}")
    
    # Update our current guess for the next loop
    x = x_new

print(f"\nFinal Answer: x = {x:.8f} found in {step} steps.")
```

Bisection method took 28 iterations and Newton-Raphson method took 5 iteration.

```python
import numpy as np
#equation
def f(x):
    return x**3 - 7*x**2 + 14*x - 5

def f_prime(x):
    # The derivative: 3x^2 - 14x + 14
    return 3*x**2 - 14*x + 14

x = (7 - np.sqrt(7)) / 3  # initial guess
step = 0
rel_error = 1.0

while rel_error >= 1e-8:
    step += 1
    
    # Calculate the next guess using the Newton-Raphson formula
    x_new = x - (f(x) / f_prime(x))
    
    # Calculate the relative error
    rel_error = abs((x_new - x) / x_new)
        
    print(f"Step {step}: x = {x_new:.8f}, Error = {rel_error:.8e}")
    
    # Update our current guess for the next loop
    x = x_new

    
print(f"\nFinal Answer: x = {x:.8f} found in {step} steps.")
```

For those values of initial guesses where derivative of function is zero, the code will go astray. For example if the initial guess $x_0 = (7 - \sqrt 7)/3$, where the derivative of the given function is zero, the code won't work. For the initial guesses, that are close to roots of derivative of the function, number of iteration  will be very high to get to the precision of $\epsilon < 10^{-8}$. For example if the initial guess $x_0 = 1.45141623, it will take 74 iteration.
