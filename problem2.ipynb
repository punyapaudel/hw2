# a

```python
#equation
def f(x):
    return x**3 - 7*x**2 + 14*x - 5

#initial guess
a = 0.0
b = 1.0
c_old = 1.0  # Placeholder to start
step = 0
rel_error = 1.0

while rel_error >= 1e-8:
    step += 1
    c = (a + b) / 2  # Find the midpoint
    
    # Calculate error (skipping the very first step)
    if step > 1:
        rel_error = abs((c - c_old) / c)
        
    print(f"Step {step}: x = {c:.8f}, Error = {rel_error:.8e}")
    
    # Update boundaries for the next loop
    if f(a) * f(c) < 0:
        b = c  # Root is in the left half
    else:
        a = c  # Root is in the right half
        
    c_old = c  # Save current midpoint for the next error calculation

print(f"\nFinal Answer: x = {c:.8f} found in {step} steps.")
```

```python
#equation
def f(x):
    return x**3 - 7*x**2 + 14*x - 5

def f_prime(x):
    # The derivative: 3x^2 - 14x + 14
    return 3*x**2 - 14*x + 14

x = 0.0  # initial guess
step = 0
rel_error = 1.0

while rel_error >= 1e-8:
    step += 1
    
    # Calculate the next guess using the Newton-Raphson formula
    x_new = x - (f(x) / f_prime(x))
    
    # Calculate the relative error
    rel_error = abs((x_new - x) / x_new)
        
    print(f"Step {step}: x = {x_new:.8f}, Error = {rel_error:.8e}")
    
    # Update our current guess for the next loop
    x = x_new

print(f"\nFinal Answer: x = {x:.8f} found in {step} steps.")
```

Bisection method took 28 iterations and Newton-Raphson method took 5 iteration.

```python
#equation
def f(x):
    return x**3 - 7*x**2 + 14*x - 5

def f_prime(x):
    # The derivative: 3x^2 - 14x + 14
    return 3*x**2 - 14*x + 14

x = 1.45141623  # initial guess
step = 0
rel_error = 1.0

while rel_error >= 1e-8:
    step += 1
    
    # Calculate the next guess using the Newton-Raphson formula
    x_new = x - (f(x) / f_prime(x))
    
    # Calculate the relative error
    rel_error = abs((x_new - x) / x_new)
        
    print(f"Step {step}: x = {x_new:.8f}, Error = {rel_error:.8e}")
    
    # Update our current guess for the next loop
    x = x_new

    
print(f"\nFinal Answer: x = {x:.8f} found in {step} steps.")
```

The code will not work if the initial guess is x_0 = 1 because Newton-Raphson gives the new guess x_1 = 0, which makes relative error |x1-x0|/x1 undefined.
Also for those value of initial guess where derivative of function is zero, the code does not work. For example if the initial guess x_0 = 1.45141623, it will take 74 iteration for the same preceision.
